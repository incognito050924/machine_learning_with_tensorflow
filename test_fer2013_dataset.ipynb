```python
import os
import numpy as np
import utils.datasets as ds
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
dataset_dir = os.path.join(os.getcwd(), 'ml_dataset', 'datasets')
x_tr, t_tr, x_val, y_val, x_ts, y_ts = ds.load_npz('fer2013', dataset_dir)
```

```python
ground_truth = ['angry', 'disgust', 'fear', 'happy', 'sad', 'surprise', 'neutral']
```

```python
loaded_cnt = x_ts.shape[0]
randidx = np.random.randint(loaded_cnt, size=10)
x_ts /= 255
for i in randidx:
    curr_img = np.reshape(x_ts[i, :], (128, 128))
    curr_label = np.argmax(y_ts[i, :])
    plt.imshow(curr_img, cmap=plt.get_cmap("gray"))
    plt.title('[%d]%s - class' % (i, ground_truth[curr_label]))
    plt.colorbar()
    plt.show()
```
